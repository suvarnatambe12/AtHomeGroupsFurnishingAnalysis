```python
def multicollinearity(df):
    vif_data = pd.DataFrame()
    vif_data['IndependVariables'] = df.columns
    vif_data['VIF'] = [variance_inflation_factor(df.values, i)\
                       for i in range(len(df.columns))]
    vif_data = vif_data.sort_values('VIF', ascending = False)
    while vif_data['VIF'].iloc[0] > 10:
        maxName = vif_data.iloc[0]
        df = df.drop([maxName.IndependVariables],axis = 1)
        vif_data = pd.DataFrame()
        vif_data['IndependVariables'] = df.columns
        vif_data['VIF'] = [variance_inflation_factor(df.values, i)\
                           for i in range(len(df.columns))]
        vif_data = vif_data.sort_values('VIF', ascending = False)
        print(vif_data)
    return df
```

```python
def backward_regression(X, y,
                           initial_list=[], 
                           threshold_in=0.01, 
                           threshold_out = 0.05, 
                           verbose=True):
    included=list(X.columns)
    while True:
        changed=False
        model = sm.OLS(y, sm.add_constant(pd.DataFrame(X[included]))).fit()
        # use all coefs except intercept
        pvalues = model.pvalues.iloc[1:]
        worst_pval = pvalues.max() # null if pvalues is empty
        if worst_pval > threshold_out:
            changed=True
            worst_feature = pvalues.idxmax()
            included.remove(worst_feature)
            if verbose:
                print('Drop {} with p-value {}'.format(worst_feature, worst_pval))
        if not changed:
            break
    return included
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
model_dtr=DecisionTreeRegressor(random_state=2021)
model_dtr.fit(X_st_train,y_train)
y_pred_dtr=model_dtr.predict(X_st_test)
print("R-Squared on train dataset={}".format(model_dtr.score(X_st_train,y_train)))
print("R-Squared on test dataset={}".format(model_dtr.score(X_st_test,y_test)))
```
